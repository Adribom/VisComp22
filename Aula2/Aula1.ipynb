## Conversão de imagem contínua

- Digitalizar espacialmente é chamado de amostragem
- Digitalizar f(x,y) é chamado de dicretização (quantização)
  - Realizar uma escala com número inteiro para a luminosidade, normalmente é utilizado 8 bits.

### Processo de Quantização de uma Imagem

Imagem é definida através de uma grandeza contínua. É necessário utilizar Variável Quantitativa Dicreta para representá-la, por isso, faz-se o processo de amostragem. 

Esse processo faz com que se obtenha uma matriz discreta e, assim, a amplitude da imagem é quantizada.

### Resolução Espacial

# Display Image

```python
import cv2
import sys
import numpy as np
```

```python
# Ler o arquivo imagem e associá-lo a uma variável
img = cv2.imread('images\cat.png')

# Exceção caso não exita o diretório informado 
if img is None:
    sys.exit("Could not read the image.")

# Mostrar a imagem em uma  janela
cv2.imshow("Janela 1", img)
k = cv2.waitKey(0) # Esperar que o usuário aperte uma tecla

# Se teclar s --> salvar imagem
# Se teclar qualquer outra, fechar janelas
if k == ord("s"):
    cv2.imwrite('images\gatin_fofo.png', img)
    cv2.destroyAllWindows()
else:
    cv2.destroyAllWindows()
```

# Desafio 1
Inverter a seguinte imagem para a posição correta:

![copa22](images\resized_copa22.png)

```python
# Ler a imagem da copa 
img_copa = cv2.imread('images\copa22.jpg')

# alterar o tamanho e a orientação da imagem 
resized_img_copa = cv2.resize(img_copa, (1280,720))
cv2.imwrite('images\\resized_copa22.png', resized_img_copa) # Salvar a imagem para ser usado na descrição Markdown acima

rotated_img_copa = cv2.rotate(resized_img_copa, cv2.ROTATE_180)

# Mostrar a imagem
cv2.imshow("Janela 1", rotated_img_copa)

# Se teclar s --> salvar imagem
# Se teclar qualquer outra, fechar janelas
if cv2.waitKey(0) == ord("s"):
    cv2.imwrite('images\\rotated_copa22.png', rotated_img_copa)
    cv2.destroyAllWindows()
else:
    cv2.destroyAllWindows()
```

# Utilização do Matplotlib 

```python
from matplotlib import pyplot as plt

# Ler imagem
img = cv2.imread('D:\Github\Geral\VisComp22\Aula2\images\cat.png')

# O Opencv automaticamente tranforma a imagem para BGR, portanto
# Converter a imagem de BGR para RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plotar a imagem com o Matplotlib
plt.imshow(img)
plt.xticks([]), plt.yticks([])  # Esconder eixo x e y
plt.show()
```
